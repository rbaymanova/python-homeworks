```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

1. Basic Plotting

```python
x = np.linspace(-10, 10)
y = x**2 - 4*x + 4

plt.plot(x, y, label='f(x) = x^2 - 4x + 4')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x^2 - 4x + 4')
plt.legend()
plt.show()
```

2. Sine and Cosine Plot

```python
x_new = np.linspace(0, 2*np.pi)
plt.plot(x_new, np.sin(x_new), 'r--o', label='sin(x)')
plt.plot(x_new, np.cos(x_new), 'b-^', label='cos(x)')
plt.legend()
plt.show()
```

4. Scatter Plot

```python
x_random = np.random.uniform(0, 10, 100)
y_random = np.random.uniform(0, 10, 100)

plt.scatter(x_random, y_random, c=np.random.rand(100), marker='o', cmap='viridis')
plt.title('Scatter Plot of Random Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.colorbar(label='Color Intensity')
plt.show()
```

5. Histogram

```python
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

6. 3D Plotting

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.cos(X**2 + Y**2)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, label='Function Value')
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("3D Surface Plot of $f(x, y) = \\cos(x^2 + y^2)$")
plt.show()
```

7. Bar Chart

```python
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [200, 150, 250, 175, 225]
colors = ['red', 'blue', 'green', 'orange', 'purple']
plt.bar(products, sales, color=colors)
plt.title('Sales Data for Products')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

8. Stacked Bar Chart

```python
time_periods = ['T1', 'T2', 'T3', 'T4']
category_a = [5, 7, 6, 8]
category_b = [3, 5, 4, 6]
category_c = [2, 3, 5, 4]

plt.bar(time_periods, category_a, label='Category A', color='skyblue')
plt.bar(time_periods, category_b, bottom=category_a, label='Category B', color='orange')
plt.bar(time_periods, category_c, bottom=np.array(category_a) + np.array(category_b), label='Category C', color='green')
plt.title('Stacked Bar Chart of Categories Over Time Periods')
plt.xlabel('Time Periods')
plt.ylabel('Values')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
